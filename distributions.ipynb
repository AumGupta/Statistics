# **Binomial, Negative Binomial, and Bernoulli Distributions**

---

## **1. Bernoulli Distribution**
- **Definition**: The Bernoulli distribution models a single trial with two possible outcomes: "success" (probability $ p $) and "failure" (probability $ 1 - p $).
- **PMF**:
  $$
  P(X = x) = 
  \begin{cases} 
      p & \text{if } x = 1 \text{ (success)} \\
      1 - p & \text{if } x = 0 \text{ (failure)}
  \end{cases}
  $$
  where $ x \in \{0, 1\} $.
- **Mean**: $ \mu = p $
- **Variance**: $ \sigma^2 = p(1 - p) $
- **Applications**: Coin toss, binary outcomes like pass/fail or yes/no.

---

## **2. Binomial Distribution**
- **Definition**: Models the number of successes ($ X $) in $ n $ independent Bernoulli trials, each with a success probability $ p $.
- **PMF**:
  $$
  P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}, \; k = 0, 1, \ldots, n
  $$
  where $ \binom{n}{k} = \frac{n!}{k!(n-k)!} $ is the number of ways to choose $ k $ successes from $ n $ trials.
- **Mean**: $ \mu = np $
- **Variance**: $ \sigma^2 = np(1 - p) $
- **Key Features**:
  - $ X $ is discrete and bounded between 0 and $ n $.
  - As $ n \to \infty $, and $ p $ remains constant, the binomial distribution approximates the **normal distribution**.
- **Applications**: Quality control, number of heads in multiple coin tosses.

---

## **3. Negative Binomial Distribution**
- **Definition**: Models the number of trials ($ X $) needed to achieve $ r $ successes, with success probability $ p $ for each trial. 
- **PMF**:
  $$
  P(X = k) = \binom{k-1}{r-1} p^r (1-p)^{k-r}, \; k \geq r
  $$
  where $ \binom{k-1}{r-1} = \frac{(k-1)!}{(r-1)!(k-r)!} $.
- **Mean**: $ \mu = \frac{r}{p} $
- **Variance**: $ \sigma^2 = \frac{r(1-p)}{p^2} $
- **Key Features**:
  - Describes overdispersed data (where variance > mean).
  - A special case ($ r = 1 $) is the **Geometric Distribution**, which models the trials required for the first success.
- **Applications**: Count data where the variance exceeds the mean (e.g., number of goals scored in sports, waiting times).

---

## **Key Differences**
| Feature              | Bernoulli          | Binomial             | Negative Binomial         |
|----------------------|--------------------|----------------------|---------------------------|
| **Trials**           | Single            | Fixed ($ n $)      | Variable (until $ r $ successes) |
| **Outcomes**         | Success/Failure   | Number of successes  | Number of trials          |
| **Mean**             | $ p $           | $ np $             | $ \frac{r}{p} $         |
| **Variance**         | $ p(1-p) $      | $ np(1-p) $        | $ \frac{r(1-p)}{p^2} $  |
| **PMF Shape**        | 2 values          | Finite values (0â€“$ n $) | Infinite for $ k \geq r $|

---

## **Quick Connections**
- **Bernoulli** is a single trial.
- **Binomial** is the sum of $ n $ Bernoulli trials.
- **Negative Binomial** extends the Bernoulli process to count trials needed for $ r $ successes.
